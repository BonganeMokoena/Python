```python
import turtle

turtle.write("Welcome to python")
turtle.penup()
turtle.goto(0,-20)
turtle.pendown()
turtle.write("Welcome to Computer Science")
turtle.penup()
turtle.goto(0,-40)
turtle.pendown()
turtle.write("Programming is fun")

turtle.done()
```

```python
#Question 1.13 (Drawing a cross)
import turtle

turtle.forward(200)
turtle.penup()
turtle.goto(100, 100)
turtle.pendown()
turtle.goto(100, -100)
turtle.right(90)
turtle.done()
```

```python
#Question 1.17 (Drawing a red line connecting points (-39, 48) and (50, -50) and displays coordinates of the points, respectively)
import turtle

turtle.color("red")
turtle.penup()
turtle.goto(-39, 48)
turtle.pendown()
turtle.write("(-39, 48)")
turtle.goto(50, -50)
turtle.write("(50, -50)")
turtle.right(45)

turtle.done()
```

```python
#Question 1.20 (drawing a rectanguloid)
import turtle

turtle.bgcolor("brown")
turtle.penup()
turtle.goto(-100, 0)
turtle.pendown()
turtle.goto(100, 0)
turtle.goto(100, -50)
turtle.goto(-100, -50)
turtle.goto(-100, 0)
turtle.goto(-75, 25)
turtle.goto(125, 25)
turtle.goto(100, 0)
turtle.penup()
turtle.goto(-75, 25)
turtle.pendown()
turtle.goto(-75, -25)
turtle.goto(125, -25)
turtle.goto(125, 25)
turtle.penup()
turtle.goto(-75, -25)
turtle.pendown()
turtle.goto(-100, -50)
turtle.penup()
turtle.goto(100, -50)
turtle.pendown()
turtle.goto(125, -25)
turtle.left(45)

turtle.done()
```

```python
#Question 1.21 (drawing a clock that shows time as 9:15:00)
import turtle

turtle.bgcolor("orange")
turtle.penup()
turtle.goto(0, -95)
turtle.pendown()
turtle.circle(100)
turtle.penup()
turtle.goto(0, 90)
turtle.pendown()
turtle.write('12')
turtle.penup()
turtle.goto(90, 0)
turtle.pendown()
turtle.write('3') 
turtle.penup()
turtle.goto(0, -90)
turtle.pendown()
turtle.write('6')
turtle.penup()
turtle.goto(-90, 0)
turtle.pendown()
turtle.write('9')

turtle.color("green") #Hours indicator
turtle.penup()
turtle.goto(5, 5)
turtle.pendown()
turtle.goto(-50, 5)

turtle.color("red")  #Minutes indicator
turtle.penup()
turtle.goto(5,5)
turtle.pendown()
turtle.goto(75, 5)

turtle.color("yellow")  #Seconds indicator
turtle.penup()
turtle.goto(5,5)
turtle.pendown()
turtle.goto(5, 80)

turtle.color("white")  #Time on the clock
turtle.penup()
turtle.goto(-15, -110)
turtle.pendown()
turtle.write("9:15:00")

turtle.done()
```
